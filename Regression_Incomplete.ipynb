```python
df = pd.read_csv('tweets_stocks_combined_final.csv')
df.head()
```

```python
x = df[['favorites','retweets', 'tweet_length', 'compound']]
y = df['5mins_price_diff_abs'].values
```

```python
print('Pearson Correlation Coefficient of variables in X\n')
print(x.corr())
```
